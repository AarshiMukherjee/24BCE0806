```python
#Importing the required libraries and loading the dataset
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np

```

```python
#Load the dataset and Display First and Last 3 rows of the dataset
df = pd.read_csv("https://raw.githubusercontent.com/salemprakash/EDA/main/Data/horse.csv")
print(df.head(3))
print(df.tail(3))
```

```python
# Check dataset dimensions
print("Number of Rows:", df.shape[0])
print("Number of Columns:", df.shape[1])
```

```python
# Summary statistics
print(df.describe())
pd.isna(df).any()
```

```python
#Check the data types of each column
df.dtypes
```

```python
#Check for null values
df.notnull()
```

```python
#Total no. of null values
df.isnull().sum()
```

```python
# Percentage of missing values per column
missing_percent = df.isnull().mean() * 100
missing_percent.sort_values(ascending=False)
```

```python
# Drop columns with more than 50% missing values
threshold = 0.5  # 50%
df = df.loc[:, df.isnull().mean() < threshold]

print("Remaining columns:", df.shape[1])
print(df)
```

```python
#Columns with missing values after dropping columns
df.isnull().sum()[df.isnull().sum() > 0]
```

```python
#Replacing missing data in both numerical and categorical data columns
numeric_cols = df.select_dtypes(include=['number']).columns
categorical_cols = df.select_dtypes(include=['object']).columns
```

```python
#Replacing numerical columns with Mean
for col in numeric_cols:
    df[col].fillna(df[col].mean(), inplace=True)
```

```python
#Replacing categorical columns with Mode
for col in categorical_cols:
    df[col].fillna(df[col].mode()[0], inplace=True)
```

```python
#Verify if data has been cleaned properly
print("Total missing values after cleaning:", df.isnull().sum().sum())
df.isnull().sum()[df.isnull().sum() > 0]
```

```python
 #Check for null values
 df.notnull()
```

```python
#Load the necessary package for 1D analysis
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
from scipy.stats import chi2_contingency
from mpl_toolkits.mplot3d import Axes3D
import plotly.graph_objects as go
```

```python
# Select a numerical column for analysis
num_col = "pulse"
```

```python
# Calculate summary statistics
mean_val = df[num_col].mean()
median_val = df[num_col].median()
mode_val = df[num_col].mode().values[0]
skewness_val = df[num_col].skew()
kurtosis_val = df[num_col].kurt()
```

```python
# Print summary statistics
print("Summary Statistics:")
print(f"Mean: {mean_val:.2f}")
print(f"Median: {median_val:.2f}")
print(f"Mode: {mode_val:.2f}")
print(f"Skewness: {skewness_val:.2f}")
print(f"Kurtosis: {kurtosis_val:.2f}")
```

```python
# Weighted Mean
pulse = df['pulse']
weights = np.ones(len(pulse))
weighted_mean = np.average(pulse, weights=weights)
print("Weighted Mean:", weighted_mean)

# Quadratic Mean (Root Mean Square)
quadratic_mean = np.sqrt(np.mean(pulse**2))
print("Quadratic Mean (RMS):", quadratic_mean)

# Harmonic Mean
harmonic_mean = stats.hmean(pulse)
print("Harmonic Mean:", harmonic_mean)

# Geometric Mean
geometric_mean = stats.gmean(pulse)
print("Geometric Mean:", geometric_mean)
```

```python
# Quadratic Mean (Root Mean Square)
quadratic_mean = np.sqrt(np.mean(pulse**2))
print("Quadratic Mean (RMS):", quadratic_mean)

# Harmonic Mean
harmonic_mean = stats.hmean(pulse)
print("Harmonic Mean:", harmonic_mean)

# Geometric Mean
geometric_mean = stats.gmean(pulse)
print("Geometric Mean:", geometric_mean)
```

```python
# Quartiles
quartiles = pulse.quantile([0.25, 0.5, 0.75])
print("Quartiles:\n", quartiles)

# Deciles
deciles = pulse.quantile(np.arange(0.1, 1.0, 0.1))
print("Deciles:\n", deciles)

# Percentiles
percentiles = pulse.quantile(np.arange(0.01, 1.0, 0.01))
print("Percentiles:\n", percentiles)

```

```python
# Range
range_pulse = pulse.max() - pulse.min()
print("Range:", range_pulse)

# Interquartile Range (IQR)
iqr = stats.iqr(pulse)
print("Interquartile Range (IQR):", iqr)

# Interdecile Range
interdecile_range = deciles.iloc[-1] - deciles.iloc[0]  # 90th - 10th
print("Interdecile Range:", interdecile_range)

# Standard Deviation
std_dev = pulse.std()
print("Standard Deviation:", std_dev)

# Variance
variance = pulse.var()
print("Variance:", variance)

```

```python

# Frequency Distribution
freq_dist = pulse.value_counts().sort_index()
print("Frequency Distribution:\n", freq_dist)

# Histogram and Frequency Polygon
plt.figure(figsize=(10,5))
counts, bins, patches = plt.hist(pulse, bins=10, edgecolor='black', alpha=0.7)
plt.plot((bins[:-1] + bins[1:]) / 2, counts, marker='o', color='red')  # frequency polygon
plt.title("Histogram & Frequency Polygon of Pulse")
plt.xlabel("Pulse")
plt.ylabel("Frequency")
plt.show()

# Relative Frequency Distribution
rel_freq = freq_dist / freq_dist.sum()
print("Relative Frequency Distribution:\n", rel_freq)

# Pie Chart (categorical bins)
bins = pd.cut(pulse, bins=5)
pie_counts = bins.value_counts()
plt.figure(figsize=(6,6))
plt.pie(pie_counts, labels=pie_counts.index.astype(str), autopct='%1.1f%%', startangle=90)
plt.title("Pulse Distribution Pie Chart")
plt.show()

# Stacked Bar Plot
bar_counts = pd.crosstab(index=bins, columns="count")
bar_counts.plot(kind='bar', stacked=True)
plt.title("Stacked Bar Plot of Pulse")
plt.ylabel("Count")
plt.show()
```
