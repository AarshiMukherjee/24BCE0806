```python
#Importing the required libraries and loading the dataset
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np

```

```python
#Load the dataset and Display First and Last 3 rows of the dataset
df = pd.read_csv("https://raw.githubusercontent.com/salemprakash/EDA/main/Data/horse.csv")
print(df.head(3))
print(df.tail(3))
```

```python
# Check dataset dimensions
print("Number of Rows:", df.shape[0])
print("Number of Columns:", df.shape[1])
```

```python
# Summary statistics
print(df.describe())
pd.isna(df).any()
```

```python
#Check the data types of each column
df.dtypes
```

```python
#Check for null values
df.notnull()
```

```python
#Total no. of null values
df.isnull().sum()
```

```python
# Percentage of missing values per column
missing_percent = df.isnull().mean() * 100
missing_percent.sort_values(ascending=False)
```

```python
# Drop columns with more than 50% missing values
threshold = 0.5  # 50%
df = df.loc[:, df.isnull().mean() < threshold]

print("Remaining columns:", df.shape[1])
print(df)
```

```python
#Columns with missing values after dropping columns
df.isnull().sum()[df.isnull().sum() > 0]
```

```python
#Replacing missing data in both numerical and categorical data columns
numeric_cols = df.select_dtypes(include=['number']).columns
categorical_cols = df.select_dtypes(include=['object']).columns
```

```python
#Replacing numerical columns with Mean
for col in numeric_cols:
    df[col].fillna(df[col].mean(), inplace=True)
```

```python
#Replacing categorical columns with Mode
for col in categorical_cols:
    df[col].fillna(df[col].mode()[0], inplace=True)
```

```python
#Verify if data has been cleaned properly
print("Total missing values after cleaning:", df.isnull().sum().sum())
df.isnull().sum()[df.isnull().sum() > 0]
```

```python
 #Check for null values
 df.notnull()
```

```python
# Plot histograms for numeric columns
for col in numeric_cols:
    plt.figure(figsize=(6, 4))
    sns.histplot(df[col], kde=True, bins=30)
    plt.title(f'Distribution of {col}')
    plt.xlabel(col)
    plt.ylabel('Frequency')
    plt.show()
```

```python
#visualisation for categorical columns
for col in categorical_cols:
    plt.figure(figsize=(6, 4))
    sns.countplot(x=df[col])
    plt.title(f'Frequency of {col}')
    plt.xticks(rotation=45)
    plt.xlabel(col)
    plt.ylabel('Count')
    plt.show()
```

```python
# Summary stats for numeric columns
for col in numeric_cols:
    print(f" {col}")
    print(f"Mean   : {df[col].mean():.2f}")
    print(f"Median : {df[col].median():.2f}")
    print(f"Mode   : {df[col].mode()[0]}")
    print("-" * 30)
```

```python
# Summary stats for categorical columns
for col in categorical_cols:
    print(f" {col}")
    print(f"Mode : {df[col].mode()[0]}")
    print("-" * 30)
```

```python
# Box plots for all numeric columns
for col in numeric_cols:
    plt.figure(figsize=(6, 4))
    sns.boxplot(x=df[col])
    plt.title(f'Box Plot of {col}')
    plt.xlabel(col)
    plt.show()
```

```python
# Correlation matrix (numeric only)
corr = df[numeric_cols].corr()

# Heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Matrix")
plt.show()
```

```python
#High Positive correlation
sns.scatterplot(data=df, x='pulse', y='respiratory_rate')
plt.title('Pulse vs Respiratory Rate')
plt.show()
np.corrcoef(df["pulse"], df["respiratory_rate"])
```

```python
#Moderate Positive correlation
sns.scatterplot(data=df, x='rectal_temp', y='pulse')
plt.title('Moderate Positive: rectal_temp vs pulse')
plt.show()
print(np.corrcoef(df['rectal_temp'], df['pulse']))
```

```python
#No correlation
sns.scatterplot(data=df, x='lesion_3', y='pulse')
plt.title('No Correlation: lesion_3 vs pulse')
plt.show()
print(np.corrcoef(df['lesion_3'], df['pulse']))
```

```python
# Moderate Negative Correlation
sns.scatterplot(data=df, x='packed_cell_volume', y='respiratory_rate')
plt.title('Moderate Negative: packed_cell_volume vs respiratory_rate')
plt.show()
print(np.corrcoef(df['packed_cell_volume'], df['respiratory_rate']))
```

```python
# High Negative Correlation
sns.scatterplot(data=df, x='total_protein', y='rectal_temp')
plt.title('High Negative: total_protein vs rectal_temp')
plt.show()
print(np.corrcoef(df['total_protein'], df['rectal_temp']))
```

```python
# Perfect Negative Correlation(simulated)
df['perfect_neg'] = -df['hospital_number']
sns.scatterplot(data=df, x='hospital_number', y='perfect_neg')
plt.title('Perfect Negative: hospital_number vs perfect_neg')
plt.show()
print(np.corrcoef(df['hospital_number'], df['perfect_neg']))
```

```python
# Perfect Positive Correlation(simulated)
df['perfect_pos'] = df['hospital_number']            # identical copy
sns.scatterplot(data=df, x='hospital_number', y='perfect_pos')
plt.title('Perfect Positive: hospital_number vs perfect_pos')
plt.show()
print(np.corrcoef(df['hospital_number'], df['perfect_pos']))
```
